```python
import random

def get_word():
    words = ["nepal","india","china","korea","japan","combodia","australia","america","bhutan","bangladesh",
             "maldhives","chili","ghana","argentina","france","portugal","germany","pakisthan",
             "dubai","saudi","chad","senegal","angola"]
    return random.choice(words)

def display_hangman(attempts):
    hangman = [
        """
          +---+
          |   |
              |
              |
              |
              |
        =========
        """,
        """
          +---+
          |   |
          O   |
              |
              |
              |
        =========
        """,
        """
          +---+
          |   |
          O   |
          |   |
              |
              |
        =========
        """,
        """
          +---+
          |   |
          O   |
         /|   |
              |
              |
        =========
        """,
        """
          +---+
          |   |
          O   |
         /|\  |
              |
              |
        =========
        """,
        """
          +---+
          |   |
          O   |
         /|\  |
         /    |
              |
        =========
        """,
        """
          +---+
          |   |
          O   |
         /|\  |
         / \  |
              |
        =========
        """
    ]
    print(hangman[6 - attempts])

def play_hangman():
    word = list(get_word())
    blank = ["_"] * len(word)
    attempts = 6
    while attempts > 0:
        print(" ".join(blank))
        display_hangman(attempts)
        guess = input("Enter your guess: ")
        if guess in word:
            for i in range(len(word)):
                if word[i] == guess:
                    blank[i] = guess
            if "_" not in blank:
                print("You win! The word was " + "".join(word))
                return
        else:
            attempts -= 1
            print("Incorrect. You have", attempts, "attempts left.")
    display_hangman(attempts)
    print("You lose! The word was " + "".join(word))

if __name__ == "__main__":
    play_hangman()
```
